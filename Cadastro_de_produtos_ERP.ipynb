```python
peca = {}
pecas = []
codigo = -1


def cadastrarpeca():
    global pecas;
    global peca;
    global codigo;

    for i in range(1):
      peca['codigo'] = codigo = codigo + 1
      peca['peca'] = input('Nome da peça: ')
      peca['fabricante'] = input('Nome do fabricante: ')
      peca['valor'] = input('Valor da peça R$: ')
      pecas.append(peca.copy())
      
      

def consultarpeca():
  global pecas
  global peca
  global codigo


  while True:
    print('----1 Consultar todas as pecas----')
    print('----2 Consutar pecas por codigo interno---- ')
    print('----3 Consultar por fabricante----')
    print('----4 Voltar----')

    rConsulta = int(input())

    if rConsulta != 1 and rConsulta != 2 and rConsulta != 3 and rConsulta != 4:
      print('----Opcao invalida----')
      continue
    
    elif rConsulta == 4:
      print('----Retornou ao menu principal----')
      break

    elif rConsulta == 1:
      #Consultar todos os produtos
      for e in pecas:
        print('-' * 20)
        for i,j in e.items():
          print(i,j)
      print('-' * 20)

    elif rConsulta == 2:
      #Consulta produto por codigo
      r = int(input('----Digite o codigo do produto----'))
      res = pecas[r]
      print ('-' * 20)
      for k,v in res.items():
        print(f'{k} : {v}')
      print('-'* 20)

    elif rConsulta == 3:
      #consultar por fabricante
      r = str(input('----Digite o nome do fabricante'))
      for peca in pecas:
        if(peca['fabricante'] == r):
          for k,v in peca.items():
            print(f'{k}:{v}')
          print('-'*20)
          
    elif rConsulta == 4:
      break

            

def removerPeca():
   r = int(input('----Digite o codigo para remover o produto----'))
   for peca in pecas:
     if(peca['codigo'] == r):
       pecas.remove(peca)
   print(f'Peça codigo {r} removida')



while True:
  print('Escolha uma opcao : ')
  print('----1 Cadastrar produto----')
  print('----2 Consultar produto----')
  print('----3 Remover produto----')
  print('----4 Encerrar programa----')

  resposta = int(input());

  if resposta != 1 and resposta != 2 and resposta != 3 and resposta != 4:
    print('Opcao inválida:');
    continue

  elif resposta == 1:
  ##cadastrar peça dentro do ERP
    print('----Voce selecionou a opcao Cadastro de produtos---- ')
    cadastrarpeca()
    continue
  
  elif resposta == 2:
  ##Consulta dentro do ERP
    print('----Voce selecionou a opcao Consulta----')
    consultarpeca()
    continue

  elif resposta == 3:
    #chama funcao remover peca
    print('----Voce selecionou a opcao Remover peca----')
    removerPeca()
    continue
  
  elif reposta == 4:
    print('---Programa encerrado---')
    break



  
```
